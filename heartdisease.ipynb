```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for obj in data:
    data[obj]=le.fit_transform(data[obj].astype(str))
data.head(8)    
```

```python
data.drop(data[data['SleepTime']>15].index, axis=0)
data.drop(data[data['PhysicalHealth']>10].index, axis=0)
data.drop(data[data['MentalHealth']>10].index, axis=0)
data.shape
```

```python
data1=data.drop(['Race'],axis=1)
data1.head(8)
```

```python
x=data1.drop(['HeartDisease'],axis=1)
x.head(8)
```

```python
x.info()
```

```python
y=data1['HeartDisease']
```

```python
y.head(8)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=45)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
r=LogisticRegression()
```

```python
reg=r.fit(x_train,y_train)
```

```python
y_pred=r.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
